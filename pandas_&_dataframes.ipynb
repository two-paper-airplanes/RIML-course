```python
import pandas as pd

# Some starter code:
df = pd.read_csv("https://raw.githubusercontent.com/iamamangosteen/pythonforbusiness/master/gdp_asia.csv", index_col="country")
rows = ['Singapore','Malaysia']
cols = ['gdpPercap_1952','gdpPercap_1957']
subset1 = df.loc[rows,cols]

rows = ['Thailand','Indonesia']
cols = ['gdpPercap_1952','gdpPercap_1957']
subset2 = df.loc[rows,cols]

rows = ['Thailand','Indonesia']
cols = ['gdpPercap_1952','gdpPercap_2007']
subset3 = df.loc[rows,cols]

# Combine the subsets into a list
subsets = [subset1, subset2, subset3]

# Concatenate the subsets along the row axis (axis=0)
combined_df = pd.concat(subsets, ignore_index=True)

# Remove duplicate rows based on the 'country' column
combined_df.drop_duplicates(subset='country', keep='first', inplace=True)

print(combined_df)
```

```python
import pandas as pd

# Some starter code:
df = pd.read_csv("https://raw.githubusercontent.com/iamamangosteen/pythonforbusiness/master/gdp_asia.csv", index_col="country")

rows = ['Singapore','Malaysia']
cols = ['gdpPercap_1952','gdpPercap_1957']
subset1 = df.loc[rows,cols]
subset1['country'] = subset1.index

rows = ['Thailand','Indonesia']
cols = ['gdpPercap_1952','gdpPercap_1957']
subset2 = df.loc[rows,cols]
subset2['country'] = subset2.index

rows = ['Thailand','Indonesia']
cols = ['gdpPercap_1952','gdpPercap_2007']
subset3 = df.loc[rows,cols]
subset3['country'] = subset3.index

# Combine the subsets into a list
subsets = [subset1, subset2, subset3]

# Concatenate the subsets along the row axis (axis=0)
combined_df = pd.concat(subsets, ignore_index=True)

# Remove duplicate rows based on the 'country' column
combined_df.drop_duplicates(subset='country', keep='first', inplace=True)

print(combined_df)
```
